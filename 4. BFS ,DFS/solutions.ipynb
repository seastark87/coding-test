```python
# dfs와 bfs # 1260
from collections import deque
import sys
# input = sys.stdin.readline

def DFS_with_adj_matrix(adj_matrix, start_node):
    visited = []
    stack = deque([start_node])
    
    while stack:
        node = stack.pop() # 노드 추출
        if node not in visited: # 방문하지 않은 노드인 경우만 작동
            # print(node) # 노드 처리
            visited.append(node) # 방문 처리
            # 인접 노드 입력
            for index, value in enumerate(adj_matrix[node][::-1]):
                if value: # 방문 여부를 따지지 않고 다 넣기
                    stack.append(N-index)
    return visited

def BFS_with_adj_matrix(adj_matrix, start_node):
    visited = []
    queue = deque([start_node])
    
    while queue:
        node = queue.popleft() # 노드 추출
        if node not in visited: # 방문하지 않은 노드인 경우만 작동
            # print(node) # 노드 처리
            visited.append(node) # 방문 처리
            # 인접 노드 입력
            for index, value in enumerate(adj_matrix[node]):
                if value: # 방문 여부를 따지지 않고 다 넣기
                    queue.append(index)
    return visited

N, M, V = map(int, input().split()) # 정점 수, 간선 수, 시작 정점
adj_matrix = [[0 for _ in range(N+1)] for _ in range(N+1)] # 인접 행렬 그래프 표현 사용
for _ in range(M):
    s, e = map(int, input().split())
    adj_matrix[s][e] = 1 # 인접 행렬 간선 입력하기, 가중치 없음
    adj_matrix[e][s] = 1 # 양방향이므로 반대도 입력

print(*DFS_with_adj_matrix(adj_matrix, V))
print(*BFS_with_adj_matrix(adj_matrix, V))

```

```python
# 연결 요소의 개수 # 11724 # dfs 이용 풀이
from collections import deque, defaultdict
import sys
# input = sys.stdin.readline

N, M = map(int, input().split()) # 정점 수, 간선 수
adj_list = defaultdict(list) # 인접 리스트 선언
for _ in range(M):
    n1, n2 = map(int, input().split())
    adj_list[n1].append(n2) # 인접 리스트 간선 입력
    adj_list[n2].append(n1) # 양방향 그래프 반영

visited = [0 for _ in range(N+1)] # 방문 여부
ans = 0 # 연결 요소의 수

while sum(visited) < N: # 방문하지 않은 노드가 있는 경우
    node = visited[1:].index(0)+1 # 방문하지 않은 노드 선택
    stack = deque([node]) # dfs stack
    while stack: # 연결 요소 전부 방문 처리
        node = stack.pop() # 노드 선택
        if visited[node] == 0: # 방문한 적 없는 노드인 경우
            visited[node] = 1 # 방문 처리
            for adj_node in adj_list[node]:
                stack.append(adj_node) # 인접 노드 추가
    ans += 1 # 연결 요소 수 증가시키기
print(ans)
```

```python
#  연결 요소의 개수 # chatGPT로 개선한 풀이
from collections import deque, defaultdict
import sys
# input = sys.stdin.readline
sys.setrecursionlimit(1000000)

def find_number_of_components(N, M, edges):
    """
    Find the number of connected components in a graph.

    Args:
    - N: int, the number of vertices in the graph.
    - M: int, the number of edges in the graph.
    - edges: List[Tuple[int, int]], a list of edges, where each edge is represented as a tuple of two vertices.

    Returns:
    - int, the number of connected components.
    """
    adj_list = defaultdict(list) # Adjacency list declaration
    for n1, n2 in edges:
        adj_list[n1].append(n2) # Add edge to the adjacency list
        adj_list[n2].append(n1) # Add edge for both directions
        
    visited = set() # Visited nodes set
    components = 0 # Number of connected components
    
    def depth_first_search(node):
        """
        Depth first search implementation for connected components.
        """
        visited.add(node)
        for adj_node in adj_list[node]:
            if adj_node not in visited:
                depth_first_search(adj_node)

    for node in range(1, N+1):
        if node not in visited:
            depth_first_search(node)
            components += 1
            
    return components

# Input reading
N, M = map(int, input().split())
edges = [tuple(map(int, input().split())) for _ in range(M)]

# Output result
print(find_number_of_components(N, M, edges))
```

```python
# 순열 사이클 # 10451 # bfs 이용 풀이
from collections import deque, defaultdict
import sys
# input = sys.stdin.readline

def get_components_num(length, permutation_list):
    adj_list = defaultdict(list)# 인접 리스트 사용, sparse matrix이고 edge를 접근해야 하므로
    for i in range(len(permutation_list)):
        adj_list[i+1].append(permutation_list[i]) # 인접 리스트 입력

    visited = set() # 방문 노드
    components = 0 # 연결 요소의 개수
    for i in range(1, length+1):
        if i not in visited: # 방문하지 않은 노드에 대하여
            queue = deque([i]) # 큐를 선언
            while queue:
                node = queue.popleft() # 노드 꺼내기
                if node not in visited: # 방문하지 않은 노드이면
                    visited.add(node) # 방문을 기록하기
                    for adj_node in adj_list[node]: # 인접 노드에 대하여
                        queue.append(adj_node)
            components += 1 # 연결 요소 개수 1 추가
    return components

n = int(input()) # test case 수
for _ in range(n):
    m = int(input()) # 순열의 개수
    a = list(map(int, input().split())) # 순열
    print(get_components_num(m, a))
```

```python
# 바이러스 # 2606
from collections import deque, defaultdict
import sys
# input = sys.stdin.readline

# 그래프, 양방향, 가중치 없음

def get_virused_num(n, e, edges):
    adj_list = defaultdict(list) # 인접 리스트 그래프 표현
    for n1, n2 in edges:
        adj_list[n1].append(n2) # 간선 입력
        adj_list[n2].append(n1) # 양방향
    queue = deque([1]) # bfs 큐 선언
    visited = set() # 방문 여부 집합 선언
    virused_num = 0 # 바이러스 걸린 컴퓨터 수 초기값 0
    while queue:
        node = queue.popleft()
        if node not in visited:
            virused_num += 1
            visited.add(node)
            for adj_node in adj_list[node]:
                queue.append(adj_node)
    return virused_num

n = int(input())
e = int(input())
edges = [tuple(map(int, input().split())) for _ in range(e)]
print(get_virused_num(n, e, edges)-1)
```

```python
# 유기농 배추 # 1012 # dfs 이용, 재귀 이용
from collections import deque, defaultdict
import sys
sys.setrecursionlimit(10**8)
# input = sys.stdin.readline

def get_worms_num(m, n, k, locations):
    # 그래프, 양방향, 가중치 없음
    adj_list = defaultdict(list) # 인접 리스트 선언
    locations = set(locations) # 좌표를 노드 key로 받음
    for location in locations:
        adj_list[location] = []
    moves = [(1, 0), # 상하좌우 이동
            (0, 1),
            (-1, 0),
            (0, -1)]
    for key, value in adj_list.items(): # 각 노드의 좌표에 대하여
        for move in moves: # 상하좌우 움직임에 대하여
            moved_location = (key[0]+move[0], key[1]+move[1]) # 이동된 자표
            if moved_location in adj_list: # 이동된 좌표가 유효한 노드인 경우
                adj_list[key].append(moved_location) # 간선을 추가

    visited = set() # 방문한 노드 집합
    def DFS(node):
        if node not in visited: # 방문하지 않은 노드인 경우
            visited.add(node) # 방문 노드로 추가
            for adj_node in adj_list[node]: # 인접 노드에 대하여
                DFS(adj_node)
    
    worms_num = 0 # 필요한 지렁이 수
    for node in adj_list.keys():
        if node not in visited:
            DFS(node)
            worms_num += 1
    return worms_num


t = int(input())
for _ in range(t):
    m, n, k = map(int, input().split()) # 밭의 가로, 세로, 배추 개수
    locations = [tuple(map(int, input().split())) for _ in range(k)] # 배추 위치 리스트
    print(get_worms_num(m, n, k, locations))
```

```python
# 결혼식 # 5567
from collections import defaultdict, deque
import sys
# input = sys.stdin.readline

def get_invitation_num(n, m, is_friends):
    invitations = set() # 초대할 사람들 집합
    friends = defaultdict(list) # 친구관계를 기록한 딕셔너리
    for a, b in is_friends:
        friends[a].append(b)
        friends[b].append(a)
    
    queue = deque([(1, 0)]) # 사람, 관계 거리를 담는 큐
    while queue:
        person, relation = queue.popleft() # 사람, 관계 거리
        invitations.add(person) # 초대 기록
        if relation <= 1: # 관계 거리가 여유가 있는 경우
            for friend in friends[person]:
                queue.append((friend, relation+1)) # 사람의 친구들도 초대 큐에 추가
    return len(invitations)-1


n = int(input())
m = int(input())
is_friends = []
for _ in range(m):
    is_friends.append(tuple(map(int, input().split())))
print(get_invitation_num(n, m, is_friends))
```

```python
# 단지번호붙이기 #2667
from collections import deque
from itertools import combinations_with_replacement
import sys
# input = sys.stdin.readline

n = int(input())
visited = [[False for i in range(n)] for j in range(n)] # 방문 여부
map = [] # 집 존재 여부
for _ in range(n):    
    map.append([int(i) for i in input()])

# 상하좌우 이동
moves = [
    (1, 0),
    (-1, 0),
    (0, 1),
    (0, -1),
    ]

num_bundle = 0
ans = []
for i, j in combinations_with_replacement(range(n), 2, ):
    if map[i][j] and not visited[i][j]: # 들리지 않은 집 있는지 확인
        queue = deque([(i, j)]) # 큐 (row, col)
        num_houses = 0
        while queue:
            row, col = queue.popleft()
            visited[row][col] = True
            num_houses += 1
            for move in moves: # 상하좌우에 대해서
                moved_row = row + move[0]
                moved_col = col + move[1]
                if moved_row < 0 or moved_row >= n or moved_col < 0 or moved_col >= n:
                    continue  # 지도 밖 제외
                if map[moved_row][moved_col] and not visited[moved_row][moved_col]:
                    visited[moved_row][moved_col] = True
                    queue.append((moved_row, moved_col)) # 집 있고 방문 안 했으면 큐에 추가
        num_bundle += 1 # 단지 수 추가
        ans.append(num_houses) # 개수 저장
            
print(num_bundle)
ans.sort()
for i in ans:
    print(i)
```

```python
# 미로 탐색 # 2178
from collections import deque
import sys
# input = sys.stdin.readline

n, m = map(int, input().split()) # 미로 row, col 길이
maze = [] # 미로 지도
for _ in range(n):
    maze.append([int(i) for i in input()])
min_travel = [[-1 for i in range(m)] for j in range(n)] # 최소 이동 거리

# bfs 구현
moves = [
    (1, 0),
    (-1, 0),
    (0, 1),
    (0, -1),
]
queue = deque([(0, 0, 1)])
while queue:
    x, y, travel = queue.popleft() # 큐에서 뽑기
    min_travel[x][y] = travel # 최소 이동 거리 입력
    if x == n-1 and y == m-1: # 답 찾으면 계산 중단
        break
    for move in moves:
        n_x = x + move[0] # x 좌표 이동
        n_y = y + move[1] # y 좌표 이동
        if n_x < 0 or n_x >= n or n_y < 0 or n_y >= m:
            continue # 지도 밖 제외
        else:
            if  min_travel[n_x][n_y] < 0 and maze[n_x][n_y]: # 아직 방문되지 않은 곳인 경우
                min_travel[n_x][n_y] = travel + 1 # 최소 이동 거리 입력
                queue.append((n_x, n_y, travel+1)) # 큐에 넣기

print(min_travel[n-1][m-1])
```

```python
# 인구이동 # 16234
# 국경선이 열리면 인구 이동은 무조건 발생한다
# 인구 이동이 멈추는 시점은 국경선이 더 이상 열리지 않는 시점(연합 개수 n**2)이다.
from collections import deque
from copy import deepcopy

moves = [
    (0, 1),
    (0, -1),
    (1, 0),
    (-1, 0),
]

# 처리해줄 일
# 방문처리, 인접 국가 처리, 연합에 넣기, 
# x,y를 포함한 union 계산 및 재배치 인구 수 계산
def bfs(x, y):
    q = deque([(x, y)]) # 시작 국가
    nation_num = 0 # 연합 국가 수 초기화
    total_people = 0 # 연합 총 인구 초기화
    while q:
        x, y = q.popleft() # 국가 뽑기
        if not visited[x][y]: 
            visited[x][y] = True
            nation_num += 1
            total_people += temp_maps[x][y]
            # 인접 연합 국가 넣기
            for dx, dy in moves:
                nx, ny = x+dx, y+dy # 새로운 x,y 좌표
                if 0<=nx<n and 0<=ny<n and not visited[nx][ny]: # 방문하지 않은 국가인 경우
                    if l <= abs(temp_maps[x][y] - temp_maps[nx][ny]) <= r:
                        q.append((nx, ny))
    return union, people_num

n, l, r = map(int, input().split())
maps = [list(map(int, input().split())) for _ in range(n)]
temp_maps = deepcopy(maps)

ans = 0
while True:
    # 연합 계산
    visited = [[False]*n for _ in range(n)]
    unions = []
    for i in range(n):
        for j in range(n):
            if not visited[i][j]:
                union, people_num = bfs(x, y)
                unions.append((union, people_num)) # 리스트로 넣는거 나중에 테스트해보기
    # 종료 조건
    if len(unions) == n**2:
        break            
    # 인구 재배치
    ans += 1
    for union, people_num in unions:
        for nation in union:
            temp_maps[nation[0]][nation[1]] = people_num
    
print(ans)


```

```python
# 스타트링크 # 5014
# 전체 층 수, 현위치, 목적지, 위로 이동 거리, 아래로 이동 거리
from collections import deque
import sys
# input = sys.stdin.readline

f, s, g, u, d = map(int, input().split())
min_travels = [-1 for _ in range(f+1)] # 최소 이동 수

# bfs로 풀기
# 넣을 때 정보 update
queue = deque([(s, 0)]) # 층 위치, 필요 이동 수
min_travels[s] = 0 # 최소 이동 수 기록
while queue:
    loc, travel = queue.popleft() # 큐에서 뽑기
    if loc == g:
        break # 필요한 값을 찾았으므로 더 이상 계산하지 않음
    # 이동 가능한 층 큐에 넣기
    up_loc = loc+u
    down_loc = loc-d
    if up_loc >0 and up_loc <= f and min_travels[up_loc] < 0:
        min_travels[up_loc] = travel+1 # 최소 이동 수 기록
        queue.append((up_loc, travel+1)) # 
    if down_loc >0 and down_loc <= f and min_travels[down_loc] < 0:
        min_travels[down_loc] = travel+1 # 최소 이동 수 기록
        queue.append((down_loc, travel+1)) #

if min_travels[g] < 0:
    print("use the stairs")
else:
    print(min_travels[g])
```

```python
# 안전 영역 #2468
from collections import deque
import sys
# input = sys.stdin.readline

n = int(input())
heights = []
for _ in range(n):
    heights.append(list(map(int, input().split())))

max_height = 0
for i in heights:
    for j in i:
        if j > max_height:
            max_height = j

ans_list = []
for k in range(max_height):
    safe_map = []
    for i in range(n):
        safe_map.append(list(map(lambda x: int(x)>k, heights[i])))

    moves = [
        (1, 0),
        (-1, 0),
        (0, 1),
        (0, -1),
        ]

    visited = [[False for i in range(n)] for j in range(n)]
    ans = 0

    for i in range(n):
        for j in range(n):
            if not visited[i][j] and safe_map[i][j]: # 방문하지 않은 안전 지역에 대해서
                ans += 1
                queue = deque([(i, j)]) # 초기 큐
                visited[i][j] = True # 큐에 들어갈 때 방문 업데이트
                while queue:
                    x, y = queue.popleft() # 큐에서 꺼냄
                    for move in moves:
                        n_x, n_y = x+move[0], y+move[1] # 이동된 위치
                        if 0<= n_x < n and 0 <= n_y < n and not visited[n_x][n_y] and safe_map[n_x][n_y]:
                            # 지도 내이며, 방문하지 않은 안전한 장소인 경우
                            visited[n_x][n_y] = True # 큐에 들어갈 때 방문 업데이트
                            queue.append((n_x, n_y)) # 큐에 넣음
    ans_list.append(ans)
print(max(ans_list))
```
