```python
#하노이 탑 #1914
import sys
# input = sys.stdin.readline

def hanoi(n, s, e):
    if n == 1:
        # base case
        print(s, e)
    else:
        # recursion call
        hanoi(n-1, s, 6-s-e)
        hanoi(1, s, e) 
        hanoi(n-1, 6-s-e, e)

n = int(input())

print(2**n-1)
if n <= 20:
    hanoi(n, 1, 3)
```

```python
# 색종이 만들기 #2630
# 중간 자르기
# 모두 같은색일 때 더 이상 자르지 않음
# 서로 다른 색이 있으면 다시 중간 자르기
# 더 이상 자를 수 없을 때 종료

import sys
# input = sys.stdin.readline

def cut(paper, length, white, blue):
    # base case
    if length == 1:
        if paper[0][0] == 1:
            return white, blue+1
        else:
            return white+1, blue
    else:
        b_num = sum([sum(row) for row in paper])
        if b_num == length**2:
            return white, blue+1
        elif b_num == 0:
            return white+1, blue
        else:
            w1, b1 = cut([row[:length//2] for row in paper[:length//2]], length//2)
            w2, b2 = cut([row[length//2:] for row in paper[:length//2]], length//2)
            w3, b3 = cut([row[length//2:] for row in paper[length//2:]], length//2)
            w4, b4 = cut([row[:length//2] for row in paper[length//2:]], length//2)
            return white+w1+w2+w3+w4, blue+b1+b2+b3+b4

length = int(input())
paper = [list(map(int, input().split())) for _ in range(length)]
white, blue = cut(paper, length, 0, 0)

print(white)
print(blue)
```

```python
#별 찍기 - 10 #2447
def draw_star(n, text, center):
    # base case
    if n == 1:
        return text
    else:
        # 비우기
        for i in range(center[0], center[0]+n//3):
            for j in range(center[1], center[1]+n//3):
                text[i][j] = ' '
        # recursion call
        text = draw_star(n//3, text, [center[0]-n//9*2, center[1]-n//9*2])
        text = draw_star(n//3, text, [center[0]-n//9*2, center[1]+n//9])
        text = draw_star(n//3, text, [center[0]-n//9*2, center[1]+n//9*4])
        text = draw_star(n//3, text, [center[0]+n//9, center[1]-n//9*2])
        text = draw_star(n//3, text, [center[0]+n//9, center[1]+n//9])
        text = draw_star(n//3, text, [center[0]+n//9, center[1]+n//9*4])
        text = draw_star(n//3, text, [center[0]+n//9*4, center[1]-n//9*2])
        text = draw_star(n//3, text, [center[0]+n//9*4, center[1]+n//9])
        text = draw_star(n//3, text, [center[0]+n//9*4, center[1]+n//9*4])
        return text
    
n = int(input())
text = [['*']*n for _ in range(n)]
text = draw_star(n, text, [n//3, n//3])
for i in text:
    print(''.join(i))
```

```python
# 피보나치 수열 전체 반환
def fibonachi(n):
    num_list = [0] * (n+1)
    num_list[1] = 1
    for i in range(2, n+1):
        num_list[i] = num_list[i-1] + num_list[i-2]
    return num_list

fibonachi(10)
```

```python
# 재귀 심화 # 피보나치 수열 반환
def fibo(n):
    # base case
    if n == 0:
        return [0]
    elif n == 1:
        return [0, 1]
    else:
        # recursive call
        a = fibo(n-1)
        return a + [a[-1]+a[-2]]

n = 10
fibo(n) 
```

```python
# 하노이 탑
N = int(input())

def move_tops(start, end, num):
    if num == 1:
        print(start, end)
        n = 1
        return n
    else:
        the_other = 6-start-end
        n = move_tops(start, the_other, num-1)
        n += 1
        print(start, end)
        n += move_tops(the_other, end, num-1)
        return n

print(2 ** N - 1)
if N <= 20:
    move_tops(1, 3, N)
```

```python
# 색종이 만들기
N = int(input())
paper = [list(map(int, input().split())) for i in range(N)]
W = 0
B = 0
    
def function(x, y, N):
    color = paper[x][y]
    for i in range(N):
        for j in range(N):
            if paper[x+i][y+j] != color:
                function(x, y, N//2)
                function(x+N//2, y, N//2)
                function(x, y+N//2, N//2)
                function(x+N//2, y+N//2, N//2)
                return
    if color == 0:
        global W
        W += 1
    else:
        global B
        B += 1
    

function(0, 0, N)
print(W)
print(B)
```

```python
# 별 찍기 - 10
# 리뷰
N = int(input())
paper = [['*' for _ in range(N)] for _ in range(N)]

def solution(x, y, N):
    if N == 3:
        paper[x+1][y+1] = ' '
    else:
        for i in range(x+N//3, x+2*N//3):
            for j in range(y+N//3, y+2*N//3):
                paper[i][j] = ' '
        solution(x, y, N//3)
        solution(x, y+N//3, N//3)
        solution(x, y+2*N//3, N//3)
        solution(x+N//3, y, N//3)
        solution(x+N//3, y+2*N//3, N//3)
        solution(x+2*N//3, y, N//3)
        solution(x+2*N//3, y+N//3, N//3)
        solution(x+2*N//3, y+2*N//3, N//3)
        
solution(0, 0, N)
for i in paper:
    print(''.join(i))
```

```python
# 시리얼 번호
import sys

N = int(sys.stdin.readline())
serial_nums = []
for _ in range(N):
    serial_nums.append(sys.stdin.readline().strip())

def sort_key_function(x):
    keys = []
    keys.append(len(x))
    s = 0
    for i in x:
        if '0' <= i <= '9':
            s += int(i)
    keys.append(s)
    keys.append(x)
    return tuple(keys)
            
serial_nums.sort(key = sort_key_function)
for i in serial_nums:
    print(i)
```
