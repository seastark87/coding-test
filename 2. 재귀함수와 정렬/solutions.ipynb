```python
# 하노이 탑
N = int(input())

def move_tops(start, end, num):
    if num == 1:
        print(start, end)
        n = 1
        return n
    else:
        the_other = 6-start-end
        n = move_tops(start, the_other, num-1)
        n += 1
        print(start, end)
        n += move_tops(the_other, end, num-1)
        return n

print(2 ** N - 1)
if N <= 20:
    move_tops(1, 3, N)
```

```python
# 색종이 만들기
N = int(input())
paper = [list(map(int, input().split())) for i in range(N)]
W = 0
B = 0
    
def function(x, y, N):
    color = paper[x][y]
    for i in range(N):
        for j in range(N):
            if paper[x+i][y+j] != color:
                function(x, y, N//2)
                function(x+N//2, y, N//2)
                function(x, y+N//2, N//2)
                function(x+N//2, y+N//2, N//2)
                return
    if color == 0:
        global W
        W += 1
    else:
        global B
        B += 1
    

function(0, 0, N)
print(W)
print(B)
```

```python
# 별 찍기 - 10
# 리뷰
N = int(input())
paper = [['*' for _ in range(N)] for _ in range(N)]

def solution(x, y, N):
    if N == 3:
        paper[x+1][y+1] = ' '
    else:
        for i in range(x+N//3, x+2*N//3):
            for j in range(y+N//3, y+2*N//3):
                paper[i][j] = ' '
        solution(x, y, N//3)
        solution(x, y+N//3, N//3)
        solution(x, y+2*N//3, N//3)
        solution(x+N//3, y, N//3)
        solution(x+N//3, y+2*N//3, N//3)
        solution(x+2*N//3, y, N//3)
        solution(x+2*N//3, y+N//3, N//3)
        solution(x+2*N//3, y+2*N//3, N//3)
        
solution(0, 0, N)
for i in paper:
    print(''.join(i))
```

```python
# 나무조각
numbers = list(map(int, input().split()))
while numbers != sorted(numbers):
    for i in range(1, len(numbers)):
        if numbers[i-1] > numbers[i]:
            save = numbers[i]
            numbers[i] = numbers[i-1]
            numbers[i-1] = save
            print(*numbers)
```

```python
# 시리얼 번호
import sys

N = int(sys.stdin.readline())
serial_nums = []
for _ in range(N):
    serial_nums.append(sys.stdin.readline().strip())

def sort_key_function(x):
    keys = []
    keys.append(len(x))
    s = 0
    for i in x:
        if '0' <= i <= '9':
            s += int(i)
    keys.append(s)
    keys.append(x)
    return tuple(keys)
            
serial_nums.sort(key = sort_key_function)
for i in serial_nums:
    print(i)
```
